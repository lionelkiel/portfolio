# Imports

```python
import numpy as np
import matplotlib.pyplot as plt
from MDSlib import simulate #Check MDSlib.py
```

# Simulating in Gas state
We know that the initial conditions of a gaseous state is given by a dimensionless temperature of $3$ and a dimensionless density of $0.3$. Using these as our initial conditions will thus force the system into a gaseous state.

```python
#Initial parameters
number_of_atoms = 4*3**3 #Number of particles
Total_simulation_time = 50 #Simulation time in dimensionless time
Time_step_length = 0.005 #Timestep length in dimensionless time
Temperature_Gas = 3 #Dimensionless temperature
Density_Gas = 0.3 #Dimensionless density
number_of_simulations = 10

Histograms_Gas = []
Pressures_Gas = []
Average_Pressures_Gas = []
for i in range(number_of_simulations):
    Position_Gas,Velocity_Gas,Force_Gas,Kinetic_Energy_Gas,Potential_Energy_Gas,Pair_correlation_histogram_Gas,Bins_Gas,Distances_Gas,Pressure_Gas = simulate(number_of_atoms, Total_simulation_time, Time_step_length, Temperature_Gas, Density_Gas)
    Histograms_Gas.append(Pair_correlation_histogram_Gas)
    Pressures_Gas.append(Pressure_Gas)
    Average_Pressures_Gas.append(np.average(Pressure_Gas))
    
#Averaging over the simulations
Average_Pair_Correlation_Gas = sum(Histograms_Gas)/number_of_simulations
Average_Pressure_Gas = sum(Average_Pressures_Gas)/number_of_simulations
Stadard_deviation_pressure_Gas = np.std(np.array(Pressures_Gas))
```

```python
#We check if the total energy is conserved for one simulation
plt.rc('font', size=15) 
plt.figure(figsize=[12,9])
plt.plot(np.arange(Total_simulation_time,step=Time_step_length), Kinetic_Energy_Gas+Potential_Energy_Gas, label = 'Kinetic + Potential Energy')
plt.plot(np.arange(Total_simulation_time,step=Time_step_length), Kinetic_Energy_Gas, label = 'Kinetic Energy')
plt.plot(np.arange(Total_simulation_time,step=Time_step_length), Potential_Energy_Gas, label = 'Potential Energy')
plt.xlabel('Time')
plt.ylabel('Energy')
plt.title('Energy in Gas state')
plt.legend()
plt.savefig('EnergyGas.pdf')
plt.show()
```

```python
#This can also be seen by checking the change of Lambda in the simulation
plt.rc('font', size=15) 
plt.figure(figsize=[12,9])
Theoretical_Kinetic_Energy = (number_of_atoms-1)*3*Temperature_Gas/2
Lambda = np.sqrt(Theoretical_Kinetic_Energy/Kinetic_Energy_Gas)
plt.plot(np.arange(Total_simulation_time,step=Time_step_length),Lambda)
plt.xlabel('Time')
plt.ylabel('Theoretial over Experimental Kinetic Energy')
plt.title('Lambda over time in Gas state')
plt.savefig('LambdaGas.pdf')
```

```python
#We look at the pair correlation function now
plt.rc('font', size=15) 
plt.figure(figsize=[12,9])
plt.bar(Bins_Gas[:-1],Average_Pair_Correlation_Gas,align='edge',width=(Bins_Gas[1]-Bins_Gas[0]))
plt.xlabel('Distance')
plt.xlim(0,6)
plt.ylabel('Pair Correlation')
plt.title('Pair Correlation of a Gas state')
plt.savefig('PairCorrelationGas.pdf')

#We also look at the Pressure
Error_in_measurement_Gas = Stadard_deviation_pressure_Gas/np.sqrt(number_of_simulations*(Total_simulation_time/Time_step_length))
print('The average pressure is', Average_Pressure_Gas, 'with standard deviation', Stadard_deviation_pressure_Gas)
print('The error in the measurement of this average is', Error_in_measurement_Gas)
```
